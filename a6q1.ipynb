```python
ds = xr.open_dataset(DATA)
```

```python
# A - Data Exploration
```

```python
# 1. Dimensions
```

```python
print("/n--- Dimensions ---")
```

```python
for dim, size in ds.dims.items():
    print(f"  {dim}: {size}")
```

```python
# 2. Main variable
```

```python
print("\n--- Variables ---")
for var in ds.data_vars:
    print(f"  Variable name : {var}")
    print(f"  Long name     : {ds[var].attrs.get('long_name', 'N/A')}")
    print(f"  Units         : {ds[var].attrs.get('units', 'N/A')}")
```

```python
#3 Spatial Resolution
```

```python
lat = ds["lat"].values
lon = ds["lon"].values
lat_res = round(float(np.diff(lat).mean()), 4)
lon_res = round(float(np.diff(lon).mean()), 4)
print(f"\n--- Spatial Resolution ---")
print(f"  Latitude  resolution : {lat_res}°")
print(f"  Longitude resolution : {lon_res}°")
print(f"  Grid size            : {len(lat)} lat × {len(lon)} lon")
 
```

```python
# 4. Institution
```

```python
print("\n--- Global Attributes (Institution / Source) ---")
for attr in ["institution", "source", "Conventions", "history"]:
    val = ds.attrs.get(attr, "N/A")
    print(f"  {attr:15s}: {val}")
```

```python
#5 & 6 Time Resolution and total length
```

```python
time = ds["time"]
print("\n--- Time Information ---")
print(f"  Start date   : {str(time.values[0])[:10]}")
print(f"  End date     : {str(time.values[-1])[:10]}")
print(f"  Total steps  : {len(time)} months")
n_years = len(time) // 12
print(f"  Total years  : ~{n_years} years (monthly resolution)")
 
print("\n--- Full Dataset Summary ---")
print(ds)
 
ds.close()
print("\n✓ Analysis complete.")
```
